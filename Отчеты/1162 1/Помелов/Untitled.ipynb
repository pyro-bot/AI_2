```python
data = data.sort_values(by='total day minutes', ascending=False)
data.head()
```

```python
columns_to_show = ['total day minutes', 'total eve minutes', 
                   'total night minutes']

data.groupby(['churn'])[columns_to_show].describe(percentiles=[])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = data['total day calls']
y = data['total day minutes']

fig, ax = plt.subplots()
# функция y1(x), синий, надпись y(x)
ax.plot(x, y, color="blue", label="голубая линия")

ax.set_xlabel("total day calls")
# подпись у вертикальной оси y
ax.set_ylabel("total day minutes")
# показывать условные обозначения
ax.legend()
```
