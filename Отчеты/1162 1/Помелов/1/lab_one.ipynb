```python
import pandas as pd
data=pd.read_csv('premier_league.csv')

data
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

len(data)
```

```python
data.head()
```

Рассмотрим на основе набора данных из лиги футболистов о результатах матча
Имеются следующие признаки:
* Домашняя команда (home_team)
* Команда гостей (away_team)
* Счет голов в домшней команде (home_goals)
* Счет голов в Курящей команде (away_goals)
* Результат матча (result)
* Сезон (season)

```python
homegoals = data['home_goals']
awaygoals = data['away_goals']
season = data['season']
```

```python
allgoals = awaygoals + homegoals
```

```python
data.loc[:,'all_goals'] = allgoals
data
```

```python
plt.hist(data['all_goals'])
plt.show()
```

```python
plt.figure(figsize=(11,6), dpi=100)
sns.set_style('darkgrid')
sns.boxplot(x='season', y='all_goals', data=data, palette='Blues')
plt.show()
```

```python
fig = plt.figure(figsize=(60,7), dpi=100)
sns.boxplot(x='home_team', y='home_goals', data=data, palette='Blues')
plt.show()

fig.savefig('home_goals.png', format = 'png', dpi =100)
```

```python
fig = plt.figure(figsize=(60,7), dpi=100)
sns.boxplot(x='away_team', y='home_goals', data=data, palette='Blues')
plt.show()

fig.savefig('away_goals.png', format = 'png', dpi =100)
```

```python
sns.violinplot(x = "all_goals", data=data)

# Отображаем
plt.show()
```

```python
fig = plt.figure(figsize=(60,7), dpi=100)
plt.subplot(2, 4, 1)
data['home_team'].value_counts().add(
        data['away_team'].value_counts(), fill_value=0
        ).plot(kind='bar', color='#4884af')
```

```python
fig = plt.figure(figsize=(60,60), dpi=100)
data['goals'] = data['home_goals'] + data['away_goals']
data['year'] = data['season'].apply(lambda season: int(season[-4:]))
regdata = data.groupby('year')['goals'].agg(np.sum)
plt.subplot(2, 4, 2)
plt.xlim(regdata.index.values[0], regdata.index.values[-1])
sns.regplot(x=regdata.index.values, y=regdata.values, ci=None)
```

```python
fig = plt.figure(figsize=(60,60), dpi=100)
team = 'Liverpool'
params = [['H', 'А', 'blue'], ['A', 'H', 'red']]
plt.subplot(2, 4, 5)
#for
teamwin = data[
        (data['home_team'] == team) & (data['result'] == params[0][0]) |
        (data['away_team'] == team) & (data['result'] == params[0][1])
        ]['year'].value_counts().sort_index()
teamwin.plot.bar(width=1, alpha=0.5, color=params[0][2])
teamlose = data[
        (data['home_team'] == team) & (data['result'] == params[1][0]) |
        (data['away_team'] == team) & (data['result'] == params[1][1])
        ]['year'].value_counts().sort_index()
teamlose.plot.bar(width=1, alpha=0.5, color=params[1][2])
```
